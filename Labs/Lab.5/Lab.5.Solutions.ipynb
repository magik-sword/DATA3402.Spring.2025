1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    def __init__(self, n = None, m = None, matrix_list = None): #n = rows, m = columns
        self.n = n
        self.m = m

        if matrix_list == None:
            self.matrix = [[0] * m for i in range(n)]
        else:
            first_row = matrix_list[0]
            for row in matrix_list:
                if len(row) != len(first_row):
                    print("Invalid matrix. All rows must have the same number of columns.")
                else:
                    self.n = len(matrix_list) 
                    self.m = len(first_row) 
                    self.matrix = matrix_list

    def __getitem__(self, position):
        if isinstance(position, tuple) and len(position)==2:
            i, j = position
            return self.matrix[i][j]
        elif isinstance(position, int):
            return self.matrix[position]
        else:
            print("Position must be in either form [i][j] or [i,j]")

    def __repr__(self):
        return repr(self.matrix)

    def assign_matrix(self, matrix2):
        if self.n == matrix2.n and self.m == matrix2.m:
            self.matrix = [row[:] for row in matrix2.matrix] 
        else:
            print("Both matrices must have the same dimensions.")
```

```python
matrix1 = Matrix(n = 2, m = 3)
matrix2 = Matrix(matrix_list = [[1,2,3], [4,5,6]])

print("2x3 Matrix using n and m:", matrix1)
print("2x3 Matrix using list:", matrix2, "\n")

print("Matrix value at matrix2[1][2]:", matrix2[1][2])
print("Matrix value at matrix2[0,2]:", matrix2[0,2], '\n')

print("Matrix 1 before:", matrix1)
matrix1.assign_matrix(matrix2)
print("Matrix 1 after:", matrix1)
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    return Matrix(matrix_list = [[float(c)] * m for i in range(n)])

def zeros(n, m):
    return constant(n, m, 0)

def ones(n, m):
    return constant(n, m, 1)

def eye(n):
    eye_matrix = [[0] * n for i in range(n)]

    for i in range(len(eye_matrix)):
        row = eye_matrix[i]
        
        for j in range(len(row)):
            if i == j:
                eye_matrix[i][j] = 1

    return eye_matrix
```

```python
print("Matrix of 5's:", constant(3, 2, 5), '\n')
print("Matrix of 0's:", zeros(3, 2), '\n')
print("Matrix of 1's:", ones(3, 2), '\n')
print("Identity Matrix:", eye(3))
```
