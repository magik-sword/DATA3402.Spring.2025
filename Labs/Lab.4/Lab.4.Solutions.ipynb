1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        self.count = 0
        self.maximum = maximum
    
    def increment(self):
        if self.count < self.maximum:
            self.count += 1
            return self.count
        else:
            return "Error, maximum value reached."
    
    def reset(self):
        self.count = 0
        print("The count has been reset to 0.")
```

```python
example = Counter(3)

for i in range(4):
    print("Count:", example.increment())

example.reset()
```

```python
import math 

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        raise NotImplementedError
    
    def perimeter(self):
        raise NotImplementedError
    
    def perimeter_points(self, points = 16):
        raise NotImplementedError

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        Shape.__init__(self, x, y)
        
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self, points = 16):  #We assume the given coordinates are for the top left corner
        x, y = self.get_x(), self.get_y()
        coordinates = []
        step = (2 * (self.__length + self.__width)) / points
        px, py = self.get_x(), self.get_y()
        
        for i in range(points):
            if px < x + self.__width and py == y:  
                px += step
                if px > x + self.__width:
                    px = x + self.__width
                    
            elif px == x + self.__width and py > y - self.__length:  
                py -= step
                if py < y - self.__length:
                    py = y - self.__length
                
            elif px > x and py == y - self.__length:  
                px -= step
                if px < x:
                    px = x
                
            elif px == x and py < y:  
                py += step
                if py > y:
                    py = y
            coordinates.append((px, py))
        
        return coordinates
        
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        Shape.__init__(self, x, y)
        
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return round(3.14159 * (self.__radius ** 2), 2)
    
    def perimeter(self):
        return round(2 * 3.14159 * self.__radius, 2)
    
    def perimeter_points(self, points = 16):
        coordinates = []
        
        for i in range(points):
            theta = 2 * 3.14159 * i / points
            
            x = self.get_x() + self.get_radius() * math.cos(theta)
            y = self.get_y() + self.get_radius() * math.sin(theta)
            coordinates.append((x, y))
        
        return coordinates
            
class Triangle(Shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        Shape.__init__(self, x, y)
        
    def get_base(self):
        return self.__base
    
    def get_height(self):
        return self.__height
    
    def area(self):
        return 0.5 * self.__base * self.__height
    
    def perimeter(self):
        hypotenuse = (self.__base ** 2 + self.__height ** 2) ** 0.5
        return self.__base + self.__height + hypotenuse
       
    def perimeter_points(self, points=16):  #Assume that the starting coordinates are for the top vertex
        x, y = self.get_x(), self.get_y()
        coordinates = []
        px, py = x, y  
        current_length = 0  
        step = self.__base + self.__height + ((self.__base ** 2 + self.__height ** 2) ** 0.5)/ points  
    
        x2, y2 = x - (self.__base / 2), y - self.__height  
        x3, y3 = x + (self.__base / 2), y - self.__height  

        for i in range(points):
            if current_length < self.__height:   
                py = y - current_length 
                px = x  
            elif current_length < self.__height + self.__base:  
                current_length -= self.__height  
                px = x2 + current_length  
                py = y2  
            else:  
                current_length -= (self.__height + self.__base)  
                ratio = current_length / ((self.__base ** 2 + self.__height ** 2) ** 0.5)
                py = y2 + (ratio * self.__height) 
                px = x3 - (ratio * (self.__base/2))
            
            coordinates.append((px, py))
            current_length += step  

        return coordinates
```

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
